```python
# Create a class with PRIVATE fields, private method and a main method. Print the fields 
# in main method. Call the private method in main method.
# Create a sub class and try to access the private fields and methods from sub class.
class MyClass:
    def __init__(self):
        self.__private_field = "Private Field"

    def __private_method(self):
        print("Private Method")

    def main_method(self):
        print(self.__private_field)
        self.__private_method()


class SubClass(MyClass):
    def access_private_elements(self):
        # Attempting to access private field and method from the superclass
        print(self.__private_field)  # Raises an AttributeError
        self.__private_method()  # Raises an AttributeError


# Create an instance of the class
obj = MyClass()

# Accessing private field and calling private method from the main method
obj.main_method()
# Output:
# Private Field
# Private Method

# Create an instance of the subclass
sub_obj = SubClass()
```

```python
# Create a class with PROTECTED fields and methods. Access these fields and methods 
# from any other class in the same package. 
# Also, Access the PROTECTED fields and methods from child class located in a different 
# package
# Access the PROTECTED fields and methods from any class in different package

class ParentClass:
    def __init__(self):
        self._protected_field = 10

    def _protected_method(self):
        print("This is a protected method.")


class ChildClass(ParentClass):
    def access_protected(self):
        self._protected_field = 20
        print("Accessed _protected_field from ChildClass:", self._protected_field)

        self._protected_method()
        print("Accessed _protected_method from ChildClass.")


# another_module.py

# from parent_module import ParentClass

class AnotherClass:
    def access_protected(self):
        parent_object = ParentClass()
        parent_object._protected_field = 30
        print("Accessed _protected_field from AnotherClass:", parent_object._protected_field)

        parent_object._protected_method()
        print("Accessed _protected_method from AnotherClass.")


# main_module.py

# from parent_module import ParentClass
# from another_module import AnotherClass
# from parent_module import ChildClass

parent_object = ParentClass()
parent_object._protected_field = 40
print("Accessed _protected_field from main_module:", parent_object._protected_field)

parent_object._protected_method()
print("Accessed _protected_method from main_module.")

child_object = ChildClass()
child_object.access_protected()

another_object = AnotherClass()
another_object.access_protected()
```

```python
# 3. Create a class with PUBLIC fields and methods. 
# Access the public methods and fields from any class in the same package or different 
# package.

# my_module.py

class MyClass:
    def __init__(self):
        self.public_field = 10

    def public_method(self):
        print("This is a public method.")


# another_module.py

# from my_module import MyClass

class AnotherClass:
    def __init__(self):
        self.obj = MyClass()

    def access_public(self):
        self.obj.public_field = 20
        print("Accessed public_field from AnotherClass:", self.obj.public_field)

        self.obj.public_method()
        print("Accessed public_method from AnotherClass.")


# main.py

# from my_module import MyClass
# from another_module import AnotherClass

obj = MyClass()
obj.public_field = 30
print("Accessed public_field from main.py:", obj.public_field)

obj.public_method()
print("Accessed public_method from main.py.")

another_obj = AnotherClass()
another_obj.access_public()
```
